Заполняем пропуски в признаке age

```python
data_age = df[df['age'] > 0]
X_age = data_age.drop(['age'], axis = 1)
y_age = data_age['age']
```

```python
X_age_train, X_age_test, y_age_train, y_age_test = train_test_split(X_age, y_age, test_size = 0.3)
model_age = DecisionTreeRegressor()
model_age.fit(X_age_train, y_age_train)
y_pred_age = model_age.predict(X_age_test)
print(mean_absolute_error(y_pred_age, y_age_test))
```

```python
ag = df.copy()
for i in range(len(ag)):
    if (np.isnan(ag['age'][i])):
        m = ag.drop(['age'], axis = 1)
        ag['age'][i] = model_age.predict(m[m.index == i])
print(ag)        
```

```python
X = df.drop(['Churn'], axis = 1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

Confusion Matrix и Accuracy

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```
