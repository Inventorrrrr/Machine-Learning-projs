```python
X = data.drop(['Sale ID', 'Profit'], axis = 1)
y = data['Profit']
```

```python
cat_ft = []
real_ft = []
for i in columnss:
    if (data[i].dtypes == 'object'):
        cat_ft.append(i)
    else:
        real_ft.append(i)
```

```python
X_dum = pd.get_dummies(X, drop_first = True)
```

```python
X_dum
```

```python
X_dum_train, X_dum_test, y_train, y_test = train_test_split(X_dum, y, test_size = 0.2)
```

```python
model = DecisionTreeRegressor()
model.fit(X_dum_train, y_train)
model
```

```python
y_pred = model.predict(X_dum_test)
```

```python
mean_absolute_error(y_pred, y_test)
```
